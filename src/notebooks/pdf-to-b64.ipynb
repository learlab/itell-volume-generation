# Convert PDF file to b64 string

For copy pasting into api queries

```python
import base64


pdf_path = "../data/input-docs/zero-to-production-ready-docker.pdf"

with open(pdf_path, "rb") as f:
    pdf_b64 = base64.b64encode(f.read()).decode("utf-8")

print(f"data:application/pdf;base64,{pdf_b64}")
```
